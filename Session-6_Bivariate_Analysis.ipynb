```python
certi_count=[]
deni_count=[]
labels=visadataset_df['continent'].unique()
for i in labels:
    con1=visadataset_df['continent']==i
    con2=visadataset_df['case_status']=='Certified'
    con3=visadataset_df['case_status']=='Denied'
    certi_con=con1 & con2
    deni_con=con1 & con3
    certi_count.append(len(visadataset_df[certi_con]))
    deni_count.append(len(visadataset_df[ deni_con]))
data=zip(certi_count,deni_count)
analysis_df=pd.DataFrame(data,columns=['Certified','Denied'],index=labels)
analysis_df
```

```python
analysis_df.plot(kind='bar')
```
