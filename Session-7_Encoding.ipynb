**np.where**
- np.where() function from numpy library is often used for conditional filtering or creating new columns based on conditions
- Filtering indices based on a condition
- Handling missing values
- It is used for binary string values
- If your column has more than two unique values(multiclass),np.where won't be suffcient

```python
visa_df['case_status']=np.where(visa_df['case_status']=='Certified',1,0)
visa_df
```

```python
cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
cat,num
```

```python
visa_df['has_job_experience'].unique()
```

```python
d1={'N':0,'Y':1}
d1
```

```python
visa_df['has_job_experience']=visa_df['has_job_experience'].map(d1)
visa_df
```

```python
visa_df['has_job_experience']=np.where(visa_df['has_job_experience']=='N',0,1)
visa_df
```

**OneHotEncoder**
- converts string catagories to multiple binary columns
- Avoids false ordering in the data
- Works for binary and multi-class
- can be combined with pipelines for ML preprocessing

```python
from sklearn.preprocessing import OneHotEncoder
labels=visa_df['continent'].unique()
dict1={labels[i]: i for i in range(len(labels))}
visa_df['continent']=visa_df['continent'].map(dict1)
visa_df
```
