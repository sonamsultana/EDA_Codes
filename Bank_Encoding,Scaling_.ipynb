### Scaling

```python
bank_df=pd.read_csv(r'C:\Users\Admin\Downloads\bank.csv',sep=';')
bank_df
cat=bank_df.select_dtypes(include='object').columns
num=bank_df.select_dtypes(exclude='object').columns
print(cat)
print(num)
```

- we have two scale methods
   - 1. Z-score
   - 2. MinMax score

<img class="XqHOTb IGEbUc" alt="Z = \frac{x - \mu}{\sigma}" src="https://www.gstatic.com/education/formulas2/553212783/en/z_score.svg" role="img" data-csiid="ek0baI6uHKm84-EP6cTn6As_14" data-atf="0">

```python
# step-1: x= balance data
# step-2: calculate mean: u
# step-3: calculate std: sigma
# step-4: Nr= step-1-step2
# step-5: Z= step-4/step-3
# step-6: create a newc column balance_z
# step-7: Draw the distribution plot for both balance and balance_z

# Observation: Even though values are modified data distribution kids
```

```python
x=bank_df['balance']
u=x.mean()
sd=x.std()
Nr=x-u
bank_df['balance_z']=Nr/sd
bank_df
```

```python
plt.figure(figsize=(8,3))
plt.subplot(1,2,1).hist(bank_df['balance'],bins=40)
plt.subplot(1,2,2).hist(bank_df['balance_z'],bins=40)
plt.show()
```

```python
# step-1: read the  day X
# step-2: calculate x_min
# step-3: calculate x_max
# step-4: Nr=X-X_min
# step-5: Dr= Xmax-Xmin
# step-6: wage_norm=Nr/Dr
# check the max(1) and min(0) values

# sklearn
# preprocessing
# MinMaxScalar
```

```python
Nr=bank_df['day']-bank_df['day'].min()
Dr=bank_df['day'].max()-bank_df['day'].min()
bank_df['day_norm']=Nr/Dr
```
